```python
a_int = input("Digite o primeiro numero: ")
b_int = input("Digite o segundo numero: ")
a = int(a_int)
b = int(b_int)

print("Resultados das Operações entre ", a, " e ", b,":")
print("Soma: ", a + b)
print("Substração: ", a - b)
print("Multiplicação: ", a * b)
print("Divisão: ", a / b)
print("Divisão inteira: ", a // b)
print("Resto da divisão: ", a % b)
print("Potência: ", a ** b)
```

```python
#Dicionario de Daados

pessoas = [
    {'nome': 'João', 'idade': 30, 'cidade': 'São Paulo'},
    {'nome': 'Maria', 'idade': 35, 'cidade': 'Rio de Janeiro'},
    {'nome': 'Pedro', 'idade': 25, 'cidade': 'Belo Horizonte'},
    {'nome': 'Ana', 'idade': 28, 'cidade': 'Porto Alegre'}
]

for pessoa in pessoas:
  print(pessoa['nome'], pessoa['idade'], pessoa['cidade'])
```
