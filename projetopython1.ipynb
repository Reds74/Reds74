```python
#Lista nome e sobrenome

pessoas = []

nome1 = input('Digite o nome da primeira pessoa: ')
snome1 = input('Digite o sobrenome da primeira pessoa: ')
pessoa1 = dict(nome = nome1, sobrenome = snome1)

nome2 = input('Digite o nome da segunda pessoa: ')
snome2 = input('Digite o nome da segunda pessoa: ')
pessoa2 = dict(nome = nome2, sobrenome = snome2)

nome3 = input('Digite o nome da terceira pessoa: ')
snome3 = input('Digite o sobrenome da terceir pessoa: ')
pessoa3 = dict(nome = nome3, sobrenome = snome3)

#adicionando a lista
pessoas.append(pessoa1)
pessoas.append(pessoa2)
pessoas.append(pessoa3)

#exibindo o nome completo de cada pessoa
print("Nome Completo:")
for pessoa in pessoas:
  print(pessoa['nome'], pessoa['sobrenome'])
```

```python
#Estruturas de repetição

import random

maximo = 10
sorteios_realizados = 0

while sorteios_realizados < maximo:
  sorteios_realizados = sorteios_realizados+1
  print(f"Sorteio {sorteios_realizados}:")
  valor_sortedo = random.randint(1, 10)
  if valor_sortedo==7:
    print("Voçê tirou 7, o jogo acabou!")
    break
  print(f"Valor sortado {valor_sortedo}. Jogue novamente.")

```

```python
#Loops de repetição

maximo = 10
atual=0

while atual < maximo:
  atual = atual+1
  print(f"Numero atual: {atual}.")
  if atual%2==0:
    continue
  print("É IMPAR")
```
